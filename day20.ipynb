```python
#Import then libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import .csv file and convert it to a DataFrame object
df = pd.read_csv(r"C:\Users\K DHANA\Downloads\Wholesale customers data.csv");
df.head()
```

```python
print(df.info())
```

```python
df.drop(['Channel','Region'],axis=1,inplace=True)
```

```python
array=df.values
```

```python
array
```

```python
stscaler = StandardScaler().fit(array)
x = stscaler.transform(array)
```

```python
x
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(20,6))
dendo = sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendogram')
plt.xlabel('Customer data')
plt.ylabel('Eucl Distance')
plt.show()
```

```python
len(set(dendo['color_list']))-1
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
group=AgglomerativeClustering(n_clusters=3)
group.fit_predict(x)
```

```python
cluster = group.fit_predict(x)
cluster.shape
```

```python
# model evalution tool
from sklearn.metrics import silhouette_score

# check how well your model is making cluster
silhouette_score(x,cluster)
```

```python
import warnings 
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans

wcss=[]
for i in range(2,11):
    kmeans=KMeans(n_clusters=4,init='k-means++',random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(2,11),wcss)
plt.title('The Elbow Method')
```

```python
group_num=pd.DataFrame(cluster,columns=['Group'])
group_num
```

```python
plt.plot(range(2,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
model = KMeans(n_clusters = 4, random_state = 309)
groups = model.fit_predict(x)
```

```python
groups
```

```python
cust_kmeans_data=pd.concat([df,group_num],axis=1)
cust_kmeans_data
```

```python
silhouette_score(x,groups)
```
